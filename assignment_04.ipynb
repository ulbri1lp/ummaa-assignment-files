```python
# Make two new empty rows for display_template and object_location

with open('temp.csv', 'r', encoding="utf-8") as f:
    reader = csv.reader(f)
    data = list(reader)

data[0].append("display_template")
for i in range(1, len(data)):
    data[i].append("")

data[0].append("object_location")
for i in range(1, len(data)):
    data[i].append("")

data[0].append("title")
for i in range(1, len(data)):
    data[i].append("")

data[0].append("format")
for i in range(1, len(data)):
    data[i].append("")

with open("temp_2.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
files = os.listdir("./Other Ethnographic")
# print(files)
images_downloaded = len(files)
images_uploaded = 0

with open("temp_2.csv", encoding="utf-8") as f:
    reader = csv.reader(f)
    with open("temp_3.csv", "w", newline="") as g:
        writer = csv.writer(g)
        for row in reader:
            for file in files:
                row[15] = row[0]
                if row[0] in file:
                    images_uploaded = images_uploaded + 1
                    row[13] = "image"
                    row[14] = "/objects/" + file
                    row[16] = "image/jpg"
                    # new_row = row[:13] + row[13] + row[14]

            writer.writerow(row)

print("Images from UMMAA: " + str(images_downloaded))
print("Images uploaded to collection builder: " + str(images_uploaded))
```

```python
# renaming headers 

inputFileName = "temp_3.csv"
outputFileName = "output.csv"

with open(inputFileName, 'r', encoding="utf-8") as inFile, open(outputFileName, 'w', newline="") as outfile:
    r = csv.reader(inFile)
    w = csv.writer(outfile)

    next(r, None)  # skip the first row from the reader, the old header
    # write new header
    w.writerow(['objectid', 'audience', 'alternateIdentifier', 'provenance', 'dateAccepted', 'extent', 'type', 'medium', 'date', 'subject/geographic', 'originInfo/place', 'culture', 'description', 'display_template', 'object_location', 'title', 'format'])

    # copy the rest
    for row in r:
        w.writerow(row)
```

```python

with open('ummaa_ethnographic.csv', encoding='utf-8') as f:
    data = csv.reader(f, delimiter=',')
    line_count_in = 0
    for row in data:
        line_count_in += 1 
    line_count_in = line_count_in - 1 #removing one line bc headers
    #print(f'\n {line_count_in} objects uploaded from UMMAA csv.')

with open('output.csv', encoding='utf-8') as f:
    data = csv.reader(f, delimiter=',')
    line_count_out = 0
    for row in data:
        line_count_out += 1
    line_count_out = line_count_out - 1 #removing one bc headers 
    #print(f'\n{line_count_out} objects uploaded to site.')

with open("log.txt", "w") as f:
    f.write("Objects imported: " + str(line_count_in) + "\n")
    f.write("Images imported: " + str(images_downloaded) + "\n")
    f.write("Objects exported: " + str(line_count_out) + "\n")
    f.write("Images exported: " + str(images_uploaded) + "\n")
```
