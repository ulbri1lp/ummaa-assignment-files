```python
# Making empty rows display_template, object_location, and format for image processing on collection builder
# Making empty row title for object display purposes on collection builder 

with open('temp.csv', 'r', encoding="utf-8") as f:
    reader = csv.reader(f)
    data = list(reader)

data[0].append("display_template")
for i in range(1, len(data)):
    data[i].append("")

data[0].append("object_location")
for i in range(1, len(data)):
    data[i].append("")

data[0].append("title")
for i in range(1, len(data)):
    data[i].append("")

data[0].append("format")
for i in range(1, len(data)):
    data[i].append("")

with open("temp_2.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
#Filling image rows with image data when applicable
#Filling title row with object id because collection builder needs items to have a title in order for them to show up on the site and none of these items have an official title so we are just filling it in with the object id

files = os.listdir("./Other Ethnographic")
images_downloaded = len(files)
images_uploaded = 0

with open("temp_2.csv", encoding="utf-8") as f:
    reader = csv.reader(f)
    with open("temp_3.csv", "w", newline="") as g:
        writer = csv.writer(g)
        for row in reader:
            for file in files:
                row[15] = row[0]
                if row[0] in file:
                    images_uploaded = images_uploaded + 1
                    row[13] = "image"
                    row[14] = "/objects/" + file
                    row[16] = "image/jpg"
                    # new_row = row[:13] + row[13] + row[14]

            writer.writerow(row)
```

```python
#Renaming headers

inputFileName = "temp_3.csv"
outputFileName = "output.csv"

with open(inputFileName, 'r', encoding="utf-8") as inFile, open(outputFileName, 'w', newline="") as outfile:
    r = csv.reader(inFile)
    w = csv.writer(outfile)

    next(r, None)
    w.writerow(['objectid', 'audience', 'alternateIdentifier', 'provenance', 'dateAccepted', 'extent', 'type', 'medium', 'date', 'subject/geographic', 'originInfo/place', 'culture', 'description', 'display_template', 'object_location', 'title', 'format'])

    for row in r:
        w.writerow(row)
```

```python

#Making log to track items and images imported vs items and images processed

with open('ummaa_ethnographic.csv', encoding='utf-8') as f:
    data = csv.reader(f, delimiter=',')
    line_count_in = 0
    for row in data:
        line_count_in += 1 
    line_count_in = line_count_in - 1 #removing one line to ignore headers row

with open('output.csv', encoding='utf-8') as f:
    data = csv.reader(f, delimiter=',')
    line_count_out = 0
    for row in data:
        line_count_out += 1
    line_count_out = line_count_out - 1 #removing one line to ignore headers row

with open("log.txt", "w") as f:
    f.write("Objects imported: " + str(line_count_in) + "\n")
    f.write("Images imported: " + str(images_downloaded) + "\n")
    f.write("Objects exported: " + str(line_count_out) + "\n")
    f.write("Images exported: " + str(images_uploaded) + "\n")
```
